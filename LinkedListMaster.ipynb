```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def insertAtBegin(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node

    def insertAtIndex(self, data, index):
        if index == 0:
            self.insertAtBegin(data)
            return

        curr = self.head
        pos = 0

        while curr is not None and pos + 1 < index:
            pos += 1
            curr = curr.next

        if curr is not None:
            new_node = Node(data)
            new_node.next = curr.next
            curr.next = new_node
        else:
            print("Index not present")

    def insertAtEnd(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        curr = self.head
        while curr.next is not None:
            curr = curr.next
        curr.next = new_node

    def updateNode(self, val, index):
        curr = self.head
        pos = 0

        while curr is not None and pos != index:
            pos += 1
            curr = curr.next

        if curr is not None:
            curr.data = val
        else:
            print("Index not present")

    def remove_first_node(self):
        if self.head is None:
            print("List is empty")
            return
        self.head = self.head.next

    def remove_last_node(self):
        if self.head is None:
            print("List is empty")
            return
        if self.head.next is None:
            self.head = None
            return

        curr = self.head
        while curr.next is not None and curr.next.next is not None:
            curr = curr.next
        curr.next = None

    def remove_at_index(self, index):
        if self.head is None:
            print("List is empty")
            return
        if index == 0:
            self.remove_first_node()
            return

        curr = self.head
        pos = 0

        while curr is not None and pos + 1 < index:
            pos += 1
            curr = curr.next

        if curr is None or curr.next is None:
            print("Index not present")
        else:
            curr.next = curr.next.next

    def remove_node(self, data):
        if self.head is None:
            print("List is empty")
            return

        # If the node to remove is the first node
        if self.head.data == data:
            self.remove_first_node()
            return

        curr = self.head
        while curr.next is not None and curr.next.data != data:
            curr = curr.next

        if curr.next is None:
            print("Node with data", data, "not found")
        else:
            curr.next = curr.next.next

    def display(self):
        curr = self.head
        while curr is not None:
            print(curr.data, end=" -> ")
            curr = curr.next
        print("None")


# Example Usage
ll = LinkedList()
ll.insertAtBegin(10)
ll.insertAtBegin(5)
ll.insertAtEnd(20)
ll.insertAtIndex(15, 2)
ll.display()  # Output: 5 -> 10 -> 15 -> 20 -> None
ll.updateNode(25, 2)
ll.display()  # Output: 5 -> 10 -> 25 -> 20 -> None
ll.remove_first_node()
ll.display()  # Output: 10 -> 25 -> 20 -> None
ll.remove_last_node()
ll.display()  # Output: 10 -> 25 -> None
ll.remove_at_index(1)
ll.display()  # Output: 10 -> None
ll.remove_node(10)
ll.display()  # Output: None
```
